```
def Where_are_orcs(link_page):
  orcs = []
  # in 
  import requests
  from bs4 import BeautifulSoup
  links = []
  URL = link_page #"http://www.framecaplib.com/lotrlib/html/chars/indices/fotr/group/thumb01/c0000000000040000.htm"
  page = requests.get(URL)

  soup = BeautifulSoup(page.content, 'html.parser')
  #soup

  for link in soup.findAll('img'):
   # print(link)
    links.append(link)
  #print('###########################')
  for i in links[8:]:
    #print(i)
    #print(i['alt'][-8:-4])
    try:
      orcs.append(int(i['alt'][-8:-4]))
    except:# KeyError, TypeError:
      pass
  return orcs
```

```
#smart boy ttt
import time
from PIL import Image
def Download_images(indexes, film, name_of_folder = 'orc/'):
  for i in indexes:
      if i < 10:
        num = '000' + str(i)
      elif i <100 and i >=10:
        num = '00'+str(i)
      elif i >=100 and i<1000:
        num = '0'+str(i)
      else:
        num = str(i)
      if i<1000:
        with open('/content/drive/MyDrive/Data_LOTR/'+ name_of_folder + str(i), "wb") as f:
            f.write(requests.get('http://www.framecaplib.com/lotrlib/images/'+film+'/'+film+ num + '.jpg').content)
      else:
        img = Image.open(requests.get('http://www.framecaplib.com/lotrlib/images/'+film+'/'+ film + num + '.jpg', stream = True).raw)
        img.save('/content/drive/MyDrive/Data_LOTR/'+ name_of_folder + str(i)+'.jpg')
      time.sleep(0.001)
```
